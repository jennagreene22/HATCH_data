```python
first_yr_data = first_yr_data.rename(columns={'Technology Name': 'technology', 'Country Name': 'country'})


combined_df = combined_df.merge(first_yr_data[['technology', 'country', 'First Year of Data']], 
                                 on=['technology', 'country'], 
                                 how='left')
combined_df = combined_df.sort_values(by='First Year of Data')

# Drop duplicates based on 'Country' and 'Technology', keeping the first occurrence (earliest year)
combined_df = combined_df.drop_duplicates(subset=['country', 'technology'], keep='first')

```

```python
def categorize_technology_timing(df):
    # Create a new column 'Country_Timing' and set all values to 'Late' initially
    df['Country_Timing'] = 'Late'
    
    # Group by 'Technology' and iterate through each group
    for tech, group in df.groupby('technology'):
        num_countries = group['country'].size
        print(int(num_countries))
        
        if(int(num_countries)) > 1:
            print('hello')
            # Check if all 'First Year of Data' values are the same for the current technology
            if group['First Year of Data'].nunique() == 1:
                # If all years are the same, set 'Early' for all rows in that technology group
                df.loc[df['technology'] == tech, 'Country_Timing'] = 'Early'
            else:
                # Find the earliest year for the current technology
                print('I have made it')
                earliest_year = group['First Year of Data'].min()
                # Update 'Country_Timing' for rows with the earliest year
                df.loc[(df['technology'] == tech) & (df['First Year of Data'] == earliest_year), 'Country_Timing'] = 'Early'
        else: 
            print('I am NaN')
            df.loc[df['technology'] == tech, 'Country_Timing'] = 'NaN'
    
    return df
```

```python
combined_df_earlylate = categorize_technology_timing(combined_df)
```
