```python
import os
import pandas as pd
from pandas import Series, DataFrame
```

Data sources:
- https://pure.iiasa.ac.at/id/eprint/10749/1/IR-13-008.pdf
- digitized from figures 16-18

```python
## Get current working directory
print("Current working directory:", os.getcwd())

## Find path to raw data (for the raw files)
raw_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'raw_data'))

## Find path to folder for saving cleaned csv
cleaned_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'cleaned_data'))


## Find path to folder for inflation
inflation_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'inflation'))
```

```python

target_inflation_file = "/A001RG3A086NBEA.xls"
target_inflation_filepath = inflation_data_path + target_inflation_file

target_file = "/cost data/healey/fig17-19 digitized.csv"
target_filepath = raw_data_path + target_file
```

```python
## adjusting for inflation from 1982 USD to 2022 USD
nipa = pd.read_excel(target_inflation_filepath, header=10)

year_list = []
for x in nipa['observation_date']:
    x = int(str(x)[:4])
    year_list.append(x)
nipa['Year'] = year_list
nipa.set_index('Year', drop=True, inplace=True)
nipa.drop(columns='observation_date', inplace=True)
nipa = nipa.transpose()

infl_1982_2022 = float(nipa[2022]/nipa[1982])
infl_1982_2022
```

```python
fgd = pd.read_csv(target_filepath, index_col=0)
fgd = fgd.transpose()
fgd.iloc[0] = fgd.iloc[0] * infl_1982_2022
fgd['Unit'] = '2022 USD/MW'
fgd['Data Source'] = 'Healey'
fgd['Spatial Scale'] = 'Global'
fgd['Country Code'] = 'World'
fgd['Country Name'] = 'World'
fgd['Metric'] = 'Price'
fgd['Technology Name'] = 'Flue Gas Desulphurization (Wet and Dry)'
fgd['ID'] = fgd['Technology Name'] + '_' + fgd['Metric'] + '_' + fgd['Country Code']
fgd.set_index('ID', drop=True, inplace=True)
fgd.columns.name = None
fgd
```

```python
output_file = 'fgd_price.csv'
output_file_path = os.path.join(cleaned_data_path, output_file)

fgd.to_csv(output_file_path)
print("Data saved to:", output_file_path)

```
