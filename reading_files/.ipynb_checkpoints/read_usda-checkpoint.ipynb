```python
# https://www.ers.usda.gov/data-products/adoption-of-genetically-engineered-crops-in-the-u-s/recent-trends-in-ge-adoption/
```

```python
usda_ers = pd.read_excel('raw data/usda.xlsx')
usda_ers.dropna(inplace=True)
usda_ers
```

```python
years = []
ht_corn = []
bt_corn = []
bt_cotton = []
ht_cotton = []
ht_soy = []

for val in usda_ers[usda_ers.columns[0]]:
    if 'Group' in val:
        years.append(val)
    if 'Line 1, HT Corn' in val:
        ht_corn.append(val)
    if 'Line 2, Bt Corn' in val:
        bt_corn.append(val)
    if 'Line 3, Bt Cotton' in val:
        bt_cotton.append(val)
    if 'Line 4, HT Cotton' in val:
        ht_cotton.append(val)
    if 'Line 5, HT Soybeans' in val:
        ht_soy.append(val)
        
```

```python
ht_corn_float = []
for val in ht_corn:
    val = val[16:-1]
    if len(val) > 0:
        val = float(val) / 100
    else:
        val = None
    ht_corn_float.append(val)

bt_corn_float = []
for val in bt_corn:
    val = val[16:-1]
    if len(val) > 0:
        val = float(val) / 100
    else:
        val = None
    bt_corn_float.append(val)

bt_cotton_float = []
for val in bt_cotton:
    val = val[18:-1]
    if len(val) > 0:
        val = float(val) / 100
    else:
        val = None
    bt_cotton_float.append(val)

ht_cotton_float = []
for val in ht_cotton:
    val = val[18:-1]
    if len(val) > 0:
        val = float(val) / 100
    else:
        val = None
    ht_cotton_float.append(val)

ht_soy_float = []
for val in ht_soy:
    val = val[20:-1]
    if len(val) > 0:
        val = float(val) / 100
    else:
        val = None
    ht_soy_float.append(val)
```

```python
# https://quickstats.nass.usda.gov/
    # Commodity: CORN, COTTON, SOYBEANS
    # Category: AREA PLANTED
    # Data Item: CORN - ACRES PLANTED, COTTON - ACRES PLANTED, SOYBEANS - ACRES PLANTED
    # Select Location: NATIONAL
    # Select Time: 
        # Year: 2023-1995
        # Period Type: Annual
        # Period: Year
        
```

```python
all_nass = pd.read_csv('raw data/74253603-4156-3F90-8F44-D546087BC8EA.csv', usecols=[1,16,19])
all_nass = all_nass.pivot(index='Data Item', columns='Year', values='Value')
all_nass.drop(columns=[1995,2023], inplace=True)
all_nass
```

```python
all_nass.loc['Pct HT Soy'] = ht_soy_float
all_nass.loc['Pct HT Cotton'] = ht_cotton_float
all_nass.loc['Pct BT Cotton'] = bt_cotton_float
all_nass.loc['Pct HT Corn'] = ht_corn_float
all_nass.loc['Pct BT Corn'] = bt_corn_float
all_nass
```

```python
corn_acres_int = []
for val in all_nass.loc['CORN - ACRES PLANTED']:
    val = val.replace(',', '')
    val = int(val)
    corn_acres_int.append(val)
    
all_nass.loc['CORN - ACRES PLANTED'] = corn_acres_int

cotton_acres_int = []
for val in all_nass.loc['COTTON - ACRES PLANTED']:
    val = val.replace(',', '')
    val = int(val)
    cotton_acres_int.append(val)
all_nass.loc['COTTON - ACRES PLANTED'] = cotton_acres_int

soy_acres_int = []
for val in all_nass.loc['SOYBEANS - ACRES PLANTED']:
    val = val.replace(',', '')
    val = int(val)
    soy_acres_int.append(val)
    
all_nass.loc['SOYBEANS - ACRES PLANTED'] = soy_acres_int
all_nass
```

```python
all_nass.loc['Acres HT Soy'] = all_nass.loc['Pct HT Soy'] * all_nass.loc['SOYBEANS - ACRES PLANTED']
all_nass.loc['Acres HT Cotton'] = all_nass.loc['Pct HT Cotton'] * all_nass.loc['COTTON - ACRES PLANTED']
all_nass.loc['Acres BT Cotton'] = all_nass.loc['Pct BT Cotton'] * all_nass.loc['COTTON - ACRES PLANTED']
all_nass.loc['Acres HT Corn'] = all_nass.loc['Pct HT Corn'] * all_nass.loc['CORN - ACRES PLANTED']
all_nass.loc['Acres BT Corn'] = all_nass.loc['Pct BT Corn'] * all_nass.loc['CORN - ACRES PLANTED']
all_nass
```

```python
all_nass.loc['Cumulative Acres HT Soy'] = all_nass.loc['Acres HT Soy'].cumsum()
all_nass.loc['Cumulative Acres HT Cotton'] = all_nass.loc['Acres HT Cotton'].cumsum()
all_nass.loc['Cumulative Acres BT Cotton'] = all_nass.loc['Acres BT Cotton'].cumsum()
all_nass.loc['Cumulative Acres HT Corn'] = all_nass.loc['Acres HT Corn'].cumsum()
all_nass.loc['Cumulative Acres BT Corn'] = all_nass.loc['Acres BT Corn'].cumsum()
all_nass
```

```python
ctc_usda = all_nass[-5:]
ctc_usda
```

```python
ctc_usda['Country Name'] = 'United States'
ctc_usda['Country Code'] = 'US'
ctc_usda['Spatial Scale'] = 'National'
ctc_usda['Data Source'] = 'USDA'
ctc_usda['Metric'] = 'Cumulative Acreage'
ctc_usda['Unit'] = 'cumulative acres planted'
ctc_usda['Technology Name'] = ['Herbicide-Tolerant Soybeans', 'Herbicide-Tolerant Cotton',
                               'Insect-Resistant Cotton', 'Herbicide-Tolerant Corn', 
                               'Insect-Resistant Corn']
ctc_usda['ID'] = ctc_usda['Technology Name'] + '_' + ctc_usda['Metric'] + '_' + ctc_usda['Country Code']
ctc_usda.set_index('ID', drop=True, inplace=True)
ctc_usda
```

```python
ctc_usda.to_csv('cleaned data/usda.csv')
```
