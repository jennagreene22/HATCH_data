```python
import pandas as pd
from pandas import Series, DataFrame
import country_converter as coco
import os
```

```python
# https://doi.org/10.1016/j.techfore.2004.11.001
# data digitized from Fig. 8 of the above article using https://apps.automeris.io/wpd/
```

```python
path = '/Users/ariana/desktop/historical_tech'
os.chdir(path)
```

```python
so2 = []
```

```python
so2_world = pd.read_excel('raw data/SCR_TimeSeries.xlsx', sheet_name = 'total', index_col = 0)
so2_world = so2_world.transpose()
so2_world['Country Name'] = 'World'
so2_world['Country Code'] = 'World'
so2_world['Spatial Scale'] = 'Global'
so2.append(so2_world)
```

```python
so2_japan = pd.read_excel('raw data/SCR_TimeSeries.xlsx', sheet_name = 'japan', index_col = 0)
so2_japan = so2_japan.transpose()
so2_japan['Country Name'] = 'Japan'
so2_japan['Country Code'] = coco.convert(names='Japan', to='iso2')
so2_japan['Spatial Scale'] = 'National'
so2.append(so2_japan)
```

```python
so2_us = pd.read_excel('raw data/SCR_TimeSeries.xlsx', sheet_name = 'usa', index_col = 0)
so2_us = so2_us.transpose()
so2_us['Country Name'] = 'United States'
so2_us['Country Code'] = coco.convert(names='United States', to='iso2')
so2_us['Spatial Scale'] = 'National'
so2.append(so2_us)
```

```python
so2_germany = pd.read_excel('raw data/SCR_TimeSeries.xlsx', sheet_name = 'germany', index_col = 0)
so2_germany = so2_germany.transpose()
so2_germany['Country Name'] = 'Germany'
so2_germany['Country Code'] = coco.convert(names='Germany', to='iso2')
so2_germany['Spatial Scale'] = 'National'
so2.append(so2_germany)
```

```python
so2 = pd.concat(so2)
so2['Unit'] = 'GWe'
so2['Data Source'] = 'Taylor, Rubin, & Hounshell'
so2['Metric'] = 'Cumulative Total Capacity'
so2['Technology Name'] = 'Wet Flue Gas Desulfurization Systems'
so2['ID'] = so2['Technology Name'] + '_' + so2['Metric'] + '_' + so2['Country Code']
so2.set_index('ID', drop=True, inplace=True)
so2
```

```python
so2.to_csv('cleaned data/wet_fgd_systems.csv')
```
