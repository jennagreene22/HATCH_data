```python
## adjusting for inflation from 2000 USD to 2022 USD
nipa = pd.read_excel('inflation/A001RG3A086NBEA.xls', header=10)

year_list = []
for x in nipa['observation_date']:
    x = int(str(x)[:4])
    year_list.append(x)
nipa['Year'] = year_list
nipa.set_index('Year', drop=True, inplace=True)
nipa.drop(columns='observation_date', inplace=True)
nipa = nipa.transpose()

infl_2000_2022 = float(nipa[2022]/nipa[2000])
infl_2000_2022
```

```python
def read_kurzweil(page_no, tech_name, unit):
    file_name = 'raw data/kurzweil/page' + str(page_no) + '.xls'
    df = pd.read_excel(file_name, header=2, usecols=[0, 1])
    df = df[:-3]
    reg_year = []
    mid_year = []
    for col in df[df.columns[0]]:
        if type(col) == float:
            mid_year.append(col)
        if type(col) == int:
            reg_year.append(col)
    if len(mid_year) != 0:
        for year in mid_year:
            four_digit_year = int(str(year)[:4])
            idx = df[df[df.columns[0]]==year].index.tolist()
            if four_digit_year in reg_year:
                df.drop(idx, inplace=True)
            else:
                df.replace({year: str(four_digit_year)}, inplace=True)
    seen = set()
    dupes = []
    for year in df[df.columns[0]]:
        if year in seen:
            dupes.append(year)
        else:
            seen.add(year)
    dup_dict = {}
    if len(dupes) != 0:
        for dup in dupes:
            dup_df = df[df[df.columns[0]]==dup]
            avg_dup_val = sum(dup_df[dup_df.columns[1]]) / len(dup_df[dup_df.columns[1]])
            dup_dict[dup] = avg_dup_val
            dup_idx = df[df[df.columns[0]]==dup].index.tolist()
            df.drop(dup_idx, inplace=True)
    df.set_index(df.columns[0], drop=True, inplace=True)
    df = df.transpose()
    for dup in dup_dict:
        df[dup] = dup_dict[dup]
    # inflation adjustment
    if tech_name=='Random Access Memory' or tech_name == 'Magnetic Data Storage':
        df.iloc[0] = df.iloc[0] * infl_2000_2022
    df['Data Source'] = 'Kurzweil'
    df['Country Name'] = 'World'
    df['Country Code'] = 'World'
    df['Spatial Scale'] = 'Global'
    df['Metric'] = 'Computing Capacity'
    df['Technology Name'] = tech_name
    df['Unit'] = unit
    df['ID'] = df['Technology Name'] + '_' + df['Metric'] + '_' + df['Country Code']
    df.set_index('ID', drop=True, inplace=True)
    df.columns.name = None
    file_name = 'cleaned data/' + tech_name + '.csv'
    df.to_csv(file_name)
    return df
```
