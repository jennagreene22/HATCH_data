```python
import os
import pandas as pd
from pandas import Series, DataFrame
import country_converter as coco
import warnings
warnings.filterwarnings("ignore")
```

Data sources:

World Bank Indicators, Indicator Code IT.NET.USER.ZS

```python
## Get current working directory
print("Current working directory:", os.getcwd())

## Find path to raw data (for the raw files)
raw_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'raw_data'))

## Find path to folder for saving cleaned csv
cleaned_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'cleaned_data'))
```

```python
internetshare_path = os.path.join(raw_data_path, 'worldbank_internetusersshare.xls')
internetshare = pd.read_excel(internetshare_path, sheet_name= 'Data', header = 3)
```

```python
columns_to_drop = ['Country Code', 'Indicator Name', 'Indicator Code']
internetshare = internetshare.drop(columns=columns_to_drop)
```

```python
internetshare
```

```python
internetshare['Data Source'] = 'World Bank Indicators'
internetshare['Technology Name'] = 'Household Internet Access'
internetshare['Spatial Scale'] = 'National'
internetshare['Metric'] = 'Share of Households'
internetshare['Unit'] = '%'
internetshare['Country Code'] = coco.convert(names=internetshare['Country Name'], to='iso2')
internetshare = internetshare.dropna(subset=['Country Code'])

```

```python
#Filter out regions that are not countries
internetshare_filtered = internetshare[internetshare['Country Code'] != 'not found']
internetshare = internetshare.merge(internetshare_filtered[['Country Code']], on='Country Code', how='inner')
internetshare
```

```python
internetshare['ID'] = internetshare['Technology Name'] + '_' + internetshare['Metric'] + '_' + internetshare['Country Code']
internetshare.set_index('ID', inplace=True, drop = False)
```

```python
output_file = 'internetshare.csv'
output_file_path = os.path.join(cleaned_data_path, output_file)

internetshare.to_csv(output_file_path, index = False)
print("Data saved to:", output_file_path)
```
