```python
# Import necessary packages
import os
import pandas as pd
from pandas import Series, DataFrame
import country_converter as coco
import warnings
warnings.filterwarnings("ignore")
```

Data sources:

Global Energy Monitor: "Global Energy Monitor, Global Gas Infrastructure Tracker, January 2024 release." 

```python
## Get current working directory
print("Current working directory:", os.getcwd())

## Find path to raw data (for the raw files)
raw_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'raw_data'))

## Find path to folder for saving cleaned csv
cleaned_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'cleaned_data'))
```

Target file path

```python
# Find and read the CSV file from the raw_data folder
target_file = 'GEM-GGIT-LNG-Terminals-2024-01.xlsx'
target_file_path = os.path.join(raw_data_path, target_file)
print(target_file_path)
```

```python
gem = pd.read_excel(target_file_path, sheet_name= 'LNG Terminals 2024-03-05', index_col=0)
gem
```

```python

# Filtering
gem = gem[(gem['FacilityType'] == 'Export') &
                 (gem['Status'] == 'Operating')]
gem
```

```python
years = list(range(1971, 2023))
gem_df = pd.DataFrame(index = gem['Country'].unique(), columns = years)
gem_df
```

```python
gem_pivot = gem.pivot_table(index = 'Country', columns = 'StartYear1', values = 'CapacityInMtpa', fill_value = 0)
gem_pivot.columns = gem_pivot.columns.astype(int)
gem_pivot = gem_pivot.reset_index()
gem_pivot
```

```python

gem_pivot['Country Code'] = coco.convert(names=gem_pivot['Country'], to='iso2')
```

```python
gem_pivot['Technology Name'] = 'Liquefied Natural Gas'
gem_pivot['Data Source'] = 'Global Energy Monitor'
gem_pivot['Unit'] = 'Mtpa'
gem_pivot['Metric'] = 'Export Capacity'
gem_pivot['Spatial Scale'] = 'National'
gem_pivot['ID'] = gem_pivot['Technology Name'] + '_' + gem_pivot['Metric'] + '_' + gem_pivot['Country Code']
gem_pivot = gem_pivot.rename(columns = {'Country': 'Country Name'})
gem_pivot.set_index('ID', inplace=True, drop = False)
gem_pivot
```

Remove USA because EIA keeps records that go back further

```python
gem_pivot = gem_pivot.drop(['Liquefied Natural Gas_Export Capacity_US'])
```

Save data and files

```python
output_file = 'gem.csv'
output_file_path = os.path.join(cleaned_data_path, output_file)

gem_pivot.to_csv(output_file_path, index = False)
print("Data saved to:", output_file_path)
```
