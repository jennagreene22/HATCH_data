```python
nukes = pd.read_csv('raw data/NorrisKristensen_NuclearWeapons_TimeSeries.csv', index_col=0, 
                    usecols=[0,1,2,3,4,5,6,7,8])
nukes = nukes.transpose()
nukes
```

```python
country_names = []
for country in nukes.index:
    country_names.append(country.title())
    
country_codes = []
for country in country_names:
    country_codes.append(coco.convert(names=country, to='iso2'))
```

```python
nukes['Country Name'] = country_names
nukes['Country Code'] = country_codes
nukes['Metric'] = 'Total Number'
nukes['Unit'] = '-'
nukes['Data Source'] = 'Norris & Kristensen'
nukes['Spatial Scale'] = 'National'
nukes['Technology Name'] = 'Nuclear Weapons'
nukes['ID'] = nukes['Technology Name'] + '_' + nukes['Metric'] + '_' + nukes['Country Code']
nukes.set_index('ID', drop=True, inplace=True)
nukes.columns.name = None
nukes
```

```python
nukes.to_csv('cleaned data/nuclear_weapons.csv')
```
