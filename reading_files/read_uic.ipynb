```python
high_speed_rail = pd.read_csv('raw data/UICAtlas_HighSpeedRail_TimeSeries.csv', index_col=0)
high_speed_rail = high_speed_rail.transpose()
high_speed_rail['Data Source'] = 'UIC'
high_speed_rail['Country Name'] = 'World'
high_speed_rail['Country Code'] = 'World'
high_speed_rail['Spatial Scale'] = 'Global'
high_speed_rail['Metric'] = 'Cumulative Total Capacity'
high_speed_rail['Technology Name'] = 'High-Speed Rail'
high_speed_rail['Unit'] = 'kilometers'
high_speed_rail['ID'] = high_speed_rail['Technology Name'] + '_' + high_speed_rail['Metric'] + '_' + high_speed_rail['Country Code']
high_speed_rail.set_index('ID', drop=True, inplace=True)
high_speed_rail.columns.name = None
high_speed_rail
```
