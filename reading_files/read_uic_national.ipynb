```python
## Get current working directory
print("Current working directory:", os.getcwd())

## Find path to raw data (for the raw files)
raw_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'raw_data'))

## Find path to folder for saving cleaned csv
cleaned_data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'cleaned_data'))
```

```python
# Find and read the CSV file from the raw_data folder
target_file = 'HighSpeedRail_Country.csv'
target_file_path = os.path.join(raw_data_path, target_file)
print(target_file_path)
```

```python
highspeedrail_df = pd.read_csv(target_file_path)
highspeedrail_df
highspeedrail_df = highspeedrail_df.replace({'China-Chinese Taipei': 'Taiwan'})
```

```python
highspeedrail_df['Distance (km)'] = pd.to_numeric(highspeedrail_df['Distance (km)'], errors='coerce')
```

years = list(range(1965, 2023))
highspeedrail_df = pd.DataFrame(index = highspeedrail_df['Country'].unique(), columns = years)
highspeedrail_df

```python
highspeedrail_df_pivot = highspeedrail_df.pivot_table(index = 'Country', columns = 'Year', values = 'Distance (km)', fill_value = 0)

```

```python
highspeedrail_df_pivot = highspeedrail_df_pivot.reset_index()
highspeedrail_df_pivot
```

```python
highspeedrail_df_pivot['Country Code'] = coco.convert(names=highspeedrail_df_pivot['Country'], to='iso2')
highspeedrail_df_pivot = highspeedrail_df_pivot[highspeedrail_df_pivot['Country Code'] != 'not found']
```

```python
highspeedrail_df_pivot['Technology Name'] = 'High Speed Rail'
highspeedrail_df_pivot['Data Source'] = 'International Union of Railways'
highspeedrail_df_pivot['Unit'] = 'km'
highspeedrail_df_pivot['Metric'] = 'Number of Units'
highspeedrail_df_pivot['Spatial Scale'] = 'National'
highspeedrail_df_pivot['ID'] = highspeedrail_df_pivot['Technology Name'] + '_' + highspeedrail_df_pivot['Metric'] + '_' + highspeedrail_df_pivot['Country Code']
highspeedrail_df_pivot = highspeedrail_df_pivot.rename(columns = {'Country': 'Country Name'})
highspeedrail_df_pivot.set_index('ID', inplace=True, drop = False)
```

```python
highspeedrail_df_pivot
```

```python
output_file = 'high_speed_rail.csv'
output_file_path = os.path.join(cleaned_data_path, output_file)

highspeedrail_df_pivot.to_csv(output_file_path, index = False)
print("Data saved to:", output_file_path)
```
